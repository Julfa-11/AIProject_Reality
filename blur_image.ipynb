```python
from google.colab import files
uploaded = files.upload()
from PIL import Image, ImageFilter
import matplotlib.pyplot as plt
im = Image.open("flower.jpg")
im1 = im.filter(ImageFilter.BoxBlur(2))
plt.imshow(im1)
plt.title("Blurred Image")
plt.axis('off')
```
